## Transaction

````python
transaction
````

- ```transaction_id``` has no use to the future work, cause there is no other transaction related table to correlate with.

````python
transaction.drop(['transaction_id'],1,inplace=True)
````

- Checking for Null values

````python
transaction.isna().sum()
````

- Considering all columns having null values ```online_order``` seems to have a distict difference.
    - so we remove null values from all except ```online_order```.

````python
drop_list = transaction.isna().sum()[transaction.isna().sum().values == 197].index
````

````python
transaction.dropna(subset=drop_list,axis = 0,inplace=True)
transaction.reset_index(drop=True,inplace=True)
transaction
````

````python
transaction.isna().sum()
````

````python
online_order_nan_list = transaction[transaction['online_order'].isna() == True].index.tolist()
````

````python
for i in online_order_nan_list:
    transaction.loc[i,'online_order'] = 'unknown'
````

````python
transaction['online_order'].value_counts()
````

- Particularly checking ```online_order``` column, every null values are explicitly meaning *unknown* value
    - So we renamed null values to **unknown**

````python
transaction.dtypes
````

Checking ata types of columns, it is highly noticable that ```transaction_date``` has inappropriate datatype

````python
transaction['transaction_date'] = pd.to_datetime(transaction['transaction_date'])
````

````python
transaction['transaction_date'].dtype
````

Replacing some values to make it to **float** type

````python
transaction['standard_cost'].str.strip()
transaction['standard_cost'] = transaction['standard_cost'].str.replace('$','')
transaction['standard_cost'] = transaction['standard_cost'].str.replace(',','')
transaction['standard_cost'] = transaction['standard_cost'].astype(float)
````

````python
transaction.dtypes
````

Initializing function to view *value_counts* of all **object** columns

````python
def viewCategory(table_name,col_name_lst):
    for col_name in col_name_lst:
        print(table_name[col_name].value_counts())
        print('')
````

````python
category_list = transaction.dtypes[transaction.dtypes == 'object'].index.tolist()
viewCategory(transaction,category_list)
````

- It seems that column ```brand``` has some unwanted strings attached to them, which are to be replaced.

````python
transaction['brand'] = transaction['brand'].str.replace('Cycles','')
transaction['brand'] = transaction['brand'].str.replace('Bicycles','')
transaction['brand'].str.strip()
````

````python
viewCategory(transaction,['brand'])
````

````python
transaction
````

## CustomerList

````python
customerList
````

````python
customerList.dtypes
````

Making ```first_name``` and ```last_name``` columns into a same **Full name**.

````python
customerList['customer_name'] = customerList['first_name'] + '-' + customerList['last_name']
customerList.drop(['last_name','first_name'], axis = 1, inplace = True)
````

````python
customerList
````

Replacing column names accordingly.

````python
customerList = customerList.rename(columns = {'past_3_years_bike_related_purchases':'related_purchases',
                                              'DOB':'birth_date',
                                              'address':'residential_address'})
````

Getting rid of unknown coulumns.

````python
customerList.drop(['Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20'],axis = 1, inplace=True)
customerList
````

````python
customerList.isna().sum()
````

Checking and dropping all null values row wise.

````python
customerList.dropna(axis = 0, inplace = True)
customerList.reset_index(drop = True,inplace=True)
````

````python
customerList.dtypes
````

while checking datatypes of the table, itseems like ```birth_date``` has irrelevent datatype

````python
customerList['birth_date'] = pd.to_datetime(customerList['birth_date'])
customerList['birth_date'].dtype
````

````python
customerList.head()
````

````python
category_list = customerList.dtypes[customerList.dtypes == 'object'].index.tolist()
viewCategory(customerList,category_list)
````

While checking value counts off all columns, ```country``` and ```deceased_indicator``` seems to has same values through out the table.

````python
customerList.drop(['country','deceased_indicator'],axis = 1, inplace=True)
````

Converting the temproary datatype of ```birth_date``` to integer format by calculating **age**.

````python
from datetime import date
today = date.today()
today = pd.to_datetime(today)
today
````

````python
customerList['age'] = ((today - customerList.birth_date)/np.timedelta64(1, 'Y'))
customerList['age'] = customerList['age'].astype(int)
customerList.drop(['birth_date'],axis = 1, inplace=True)
````

rounding off the remaining columns and rearranging the table.

````python
customerList['Value'] = customerList['Value'].round(3)
````

````python
customerList = customerList[['customer_name','age','gender','related_purchases','job_title','job_industry_category',
                             'wealth_segment','owns_car','tenure','property_valuation','Rank','Value','residential_address','postcode','state']]
````
